```python
name_algorithm = 'nsgaii'
name_problem   = 'wfg1' 
name_oprt      = 'baseline'
```

```python
def generational_optimas(name_algorithm, name_problem, name_oprt):
    Mega_X =  np.load('Solutions/'+name_algorithm+'/'+name_problem+'_'+name_oprt+'_X.npy')
    Mega_F =  np.load('Solutions/'+name_algorithm+'/'+name_problem+'_'+name_oprt+'_F.npy')
    n_ejec, n_gen, __, __ = Mega_X.shape
    n_ejec=1
    optimal_results_F = []
    optimal_results_X = []
    for i in tqdm(range(n_ejec)): 
      F_opt = None
      X_opt = None
      for j in range(n_gen): 
        F_opt, X_opt = archieving_strategies.get_best_opt(Mega_F[i,j], Mega_X[i,j], F_opt, X_opt)
        optimal_results_F.append(F_opt[:])
        optimal_results_X.append(X_opt[:])
    return optimal_results_F, optimal_results_X
```

```python
optimal_F, optimal_X = generational_optimas(name_algorithm, name_problem, name_oprt)
```
